```python
# Dependencies

import pandas as pd
import requests
import json

# Retrieve Google API key from config.py

from config import gkey

locations = pd.read_csv("Interim Data Files/State and County Info.csv")

locations['Latitude'] = ""
locations['Longitude'] = ""

# Build URL using the Google MAps API

base_url = "https://maps.googleapis.com/maps/api/geocode/json?"

location=""

params={
    "key": gkey
}

# iterate through datafram and retrieve lat and lng info

for index, row in locations.iterrows():

    location = row['Geography']
    params["address"] = location

    response = requests.get(base_url,params=params).json()
    
    # extract results
    
    results = response['results']
    
    try:
        print(f"Retrieving Results for Index {index}: {location}.")
        
        locations.loc[index, 'Latitude'] = results[0]['geometry']['location']['lat']
        locations.loc[index, 'Longitude'] = results[0]['geometry']['location']['lng']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```
