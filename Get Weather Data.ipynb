```python
# Dependencies and Setup
import pandas as pd
import requests
import time
import json
import base64

# Import API keys
from config import consumer_key, consumer_secret

# generate access token for API
credential = consumer_key+":"+consumer_secret

hashed_credential = base64.b64encode(bytes(credential,'utf-8')).decode('utf-8')

print(hashed_credential)

token_header = {"Content-Type": "application/x-www-form-urlencoded",
"Authorization": "Basic " + str(hashed_credential)}

token_body = "grant_type=client_credentials"

token_url = "https://api.awhere.com/oauth/token"

token_response = requests.post(token_url,headers=token_header, data = token_body).json()

print(token_response)

# store acess token
access_token = token_response["access_token"]
```

```python
locations = pd.read_csv("Interim Data Files/State and County Info with Lat and Lng.csv")

locations['Avg Temperature'] = ""
locations['Max Temperature'] = ""
locations['Min Temperature'] =  ""
locations['Avg Percipitation'] = ""
locations['Min Humidity'] = ""
locations['Max Humidity'] = ""
locations['Avg Wind'] = ""

url = "https://api.awhere.com/v2/weather/locations"

lat = ""
lng = ""

headers = {"Authorization": "Bearer " + access_token}

#loop through rows, make API calls and store data

for index, row in locations.iterrows():
    lat = row['Latitude']
    lng = row['Longitude']

    query_url = f"{url}/{lat},{lng}/norms/01-01/years/2015,2018/?&units=usa"
    
    response = requests.get(query_url, headers=headers)
    response_json = response.json()
    
    try:
        print(f"Retrieving Results for Index {index}")
        
        locations.loc[index, 'Avg Temperature'] = response_json['meanTemp']['average']
        locations.loc[index, 'Max Temperature'] = response_json['maxTemp']['average']
        locations.loc[index, 'Min Temperature'] = response_json['minTemp']['average']
        locations.loc[index, 'Avg Percipitation'] = response_json['precipitation']['average']
        locations.loc[index, 'Min Humidity'] = response_json['minHumidity']['average']
        locations.loc[index, 'Max Humidity'] = response_json['maxHumidity']['average']
        locations.loc[index, 'Avg Wind'] = response_json['averageWind']['average']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
    
```
